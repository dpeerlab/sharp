## Reads Count

```python
import numpy as np
import scipy.io
```

```python
mtx = scipy.io.mmread(
    os.path.join(path_data, "reads/matrix.mtx.gz")
)
```

```python
barcodes = pd.read_csv(
    os.path.join(path_data, "reads/barcodes.tsv.gz"),
    sep="\t", 
    compression="gzip",
    header=None
)
```

```python
features = pd.read_csv(
    os.path.join(path_data, "reads/features.tsv.gz"),
    sep="\t",
    compression="gzip",
    header=None
)
```

```python
df_demux = pd.DataFrame(
    mtx.T.todense(),
    index=barcodes[0],
    columns=features[0]
)
```

```python
df_demux
```

### Overall Stats

## Plots

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams["font.family"] = "Sathu"
```

### Histogram (sum of reads across hashtags)

```python
fig = plt.figure(figsize=(6,6), facecolor="white")
ax = fig.add_subplot(111)

ax.hist(df_demux.sum(axis=1).sort_values(ascending=False), bins=50)

ax.set_xlabel("Reads", fontweight="bold")
ax.set_ylabel("Frequency (Log)", fontweight="bold")
ax.set_title("Histogram (sum of reads across hashtags)", fontweight="bold")

plt.yscale("log")
plt.show()
```

### Histogram (sum of reads per hashtag)

```python
fig = plt.figure(figsize=(6, 6), facecolor="white")

ax = fig.add_subplot(111)
    
for idx, col in enumerate(df_demux.columns[:-1]):
    
    df_tmp = df_demux.iloc[:,idx].sort_values(ascending=False)

    ax.hist(df_tmp, bins=50, alpha=0.7)

ax.set_xlabel("Reads", fontweight="bold")
ax.set_ylabel("Frequency (Log)", fontweight="bold")
ax.tick_params(axis="x", rotation=45)
ax.set_title("Histogram (sum of reads per hashtag)", fontweight="bold")
ax.grid(b=True, linestyle="--")
ax.legend()

plt.yscale("log")
plt.tight_layout()
plt.show()
```
